```python
from sklearn.model_selection import train_test_split
X = df.drop(columns = ['outcome']);
y = df['outcome']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 1)
```

```python

y.shape
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
#1st layer 8 input so input_dim = 8 and activation function which is used is relu
# 12 repersent number of nodes in the hidden layer 1
model.add(Dense(12,input_dim = 8,activation = 'relu'))
#2nd layer ....for next layer 8 nodes is present
model.add(Dense(8,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy',optimizer = 'Adam')  # it is a binary classification so binary_crossentropy
```

```python
model.fit(X_train_scaled,y_train,epochs=100)
```

```python
y_log = model.predict(X_test_scaled)
y_log
```

```python
y_pred = np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
